```python
# 7. RELACIÓN ENTRE VARIABLES Y REPROBACIÓN
print("="*60)
print("RELACIÓN VARIABLES VS REPROBACIÓN")
print("="*60)

# Encontrar la variable de reprobación/resultado académico
columnas_resultado = [col for col in df.columns if any(keyword in col.lower() 
                     for keyword in ['reprob', 'aprobo', 'resultado', 'estado', 'calificacion'])]

if columnas_resultado:
    var_objetivo = columnas_resultado[0]  # Tomar la primera
    print(f"Variable objetivo seleccionada: {var_objetivo}")
    print(f"Valores únicos: {df[var_objetivo].unique()}")
    
    # CONVERSIÓN A NUMÉRICO PARA CORRELACIONES
    # Crear una copia de la variable objetivo convertida a numérico
    if df[var_objetivo].dtype == 'object':
        # Mapear valores categóricos a numéricos
        if set(df[var_objetivo].unique()).issubset({'AP', 'RP'}):
            df_temp = df.copy()
            df_temp[f'{var_objetivo}_num'] = df_temp[var_objetivo].map({'AP': 1, 'RP': 0})
            var_objetivo_num = f'{var_objetivo}_num'
            print(f"Conversión: AP=1 (Aprobado), RP=0 (Reprobado)")
        else:
            # Para otros casos, usar factorize
            df_temp = df.copy()
            df_temp[f'{var_objetivo}_num'] = pd.factorize(df_temp[var_objetivo])[0]
            var_objetivo_num = f'{var_objetivo}_num'
            print(df_temp[var_objetivo].unique())
            print(df_temp[var_objetivo_num].unique())
            print(f"Conversión numérica automática aplicada")
    else:
        df_temp = df.copy()
        var_objetivo_num = var_objetivo
    
    # Análisis bivariado con variables numéricas
    variables_numericas = df_temp.select_dtypes(include=[np.number]).columns.tolist()
    if var_objetivo_num in variables_numericas:
        variables_numericas.remove(var_objetivo_num)
    
    # Correlaciones con la variable objetivo (ahora numérica)
    if len(variables_numericas) > 0:
        print(f"\nCorrelaciones con {var_objetivo} (convertido a numérico):")
        correlaciones = df_temp[variables_numericas + [var_objetivo_num]].corr()[var_objetivo_num].sort_values(ascending=False)
        correlaciones_filtradas = correlaciones.drop(var_objetivo_num)
        print(correlaciones_filtradas)
        
        # Visualizar las correlaciones más importantes
        if len(correlaciones_filtradas) > 0:
            top_corr = correlaciones_filtradas.abs().sort_values(ascending=False)[:8]
            plt.figure(figsize=(12, 6))
            colors = ['red' if x < 0 else 'green' for x in correlaciones_filtradas[top_corr.index]]
            top_corr.plot(kind='bar', color=colors, alpha=0.7)
            plt.title(f'Correlaciones más fuertes con {var_objetivo}')
            plt.ylabel('Correlación (valor absoluto)')
            plt.xlabel('Variables')
            plt.xticks(rotation=45, ha='right')
            plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
            plt.tight_layout()
            plt.show()
    
    # Análisis categórico vs variable objetivo
    variables_categoricas = df.select_dtypes(include=['object', 'category']).columns.tolist()
    if var_objetivo in variables_categoricas:
        variables_categoricas.remove(var_objetivo)
        
    if len(variables_categoricas) > 0:
        print(f"\nAnálisis categórico vs {var_objetivo}:")
        n_vars = min(6, len(variables_categoricas))
        if n_vars > 0:
            plt.figure(figsize=(15, 10))
            for i, var_cat in enumerate(variables_categoricas[:n_vars], 1):
                plt.subplot(2, 3, i)
                tabla_cruzada = pd.crosstab(df[var_cat], df[var_objetivo], normalize='index') * 100
                tabla_cruzada.plot(kind='bar', ax=plt.gca(), color=['lightgreen', 'lightcoral'])
                plt.title(f'{var_cat} vs {var_objetivo}')
                plt.ylabel('Porcentaje')
                plt.xticks(rotation=90)
                plt.legend(title=var_objetivo)
            plt.tight_layout()
            plt.show()
            
            # Mostrar tabla de contingencia para la primera variable categórica
            print(f"\nTabla de contingencia - {variables_categoricas[0]} vs {var_objetivo}:")
            tabla = pd.crosstab(df[variables_categoricas[0]], df[var_objetivo], margins=True)
            print(tabla)
    
    # Estadísticas descriptivas por grupo
    print(f"\nEstadísticas descriptivas por {var_objetivo}:")
    for valor in df[var_objetivo].unique():
        print(f"\n--- {valor} ---")
        subset = df[df[var_objetivo] == valor]
        print(f"Cantidad: {len(subset)}")
        if len(variables_numericas) > 0:
            print("Promedios de variables numéricas:")
            promedios = subset[variables_numericas].mean()
            print(promedios.round(2))
            
else:
    print("No se encontró variable objetivo clara para el análisis")
```

```python
# 8. DETECCIÓN DE VALORES ATÍPICOS (OUTLIERS)
print("="*60)
print("DETECCIÓN DE VALORES ATÍPICOS")
print("="*60)

variables_numericas = df.select_dtypes(include=[np.number]).columns.tolist()

if len(variables_numericas) > 0:
    # Detectar outliers usando el método IQR
    outliers_summary = {}
    for var in variables_numericas[:8]:  # Analizar las primeras 8 variables
        Q1 = df[var].quantile(0.25)
        Q3 = df[var].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        outliers = df[(df[var] < lower_bound) | (df[var] > upper_bound)]
        outliers_summary[var] = len(outliers)
        
        print(f"{var}: {len(outliers)} outliers detectados ({len(outliers)/len(df)*100:.1f}%)")
    
    # Visualización de boxplots para detectar outliers
    plt.figure(figsize=(15, 10))
    for i, var in enumerate(variables_numericas[:8], 1):
        plt.subplot(2, 4, i)
        plt.boxplot(df[var].dropna())
        plt.title(f'Boxplot de {var}')
        plt.ylabel(var)
    plt.tight_layout()
    plt.show()
    
    # Resumen de outliers
    outliers_df = pd.DataFrame(list(outliers_summary.items()), 
                              columns=['Variable', 'Número_Outliers'])
    outliers_df['Porcentaje'] = (outliers_df['Número_Outliers'] / len(df)) * 100
    print("\nResumen de outliers por variable:")
    print(outliers_df.round(2))
```

##### ANÁLISIS EXPLORATORIO DE TASA DE REPROBACIÓN

En esta sección analizaremos las variables clave del dataset de tasa de reprobación: **REPROBADOS**, **REGISTRADOS** y **TASA_RP**

```python
# ANÁLISIS EXPLORATORIO DE TASA DE REPROBACIÓN
print("="*80)
print("ANÁLISIS DE TASA DE REPROBACIÓN - VARIABLES CLAVE")
print("="*80)

# Crear una copia para trabajar
f_tasa_reprobacion = df_tasa_reprobacion.copy()

# 1. INFORMACIÓN GENERAL
print("\n📊 INFORMACIÓN GENERAL DEL DATASET:")
print(f"• Dimensiones: {f_tasa_reprobacion.shape}")
print(f"• Número de registros (materias): {f_tasa_reprobacion.shape[0]}")
print(f"• Número de variables: {f_tasa_reprobacion.shape[1]}")

# Verificar las columnas de interés
columnas_interes = ['REPROBADOS', 'REGISTRADOS', 'TASA_RP']
print(f"\n✓ Verificación de columnas de interés:")
for col in columnas_interes:
    if col in f_tasa_reprobacion.columns:
        print(f"  • {col}: ✓ Presente")
    else:
        print(f"  • {col}: ✗ No encontrada")

# Mostrar todas las columnas disponibles
print(f"\nColumnas disponibles en el dataset:")
for i, col in enumerate(f_tasa_reprobacion.columns, 1):
    print(f"  {i}. {col}")

# Verificar tipos de datos
print(f"\nTipos de datos para variables de interés:")
for col in columnas_interes:
    if col in f_tasa_reprobacion.columns:
        print(f"  • {col}: {f_tasa_reprobacion[col].dtype}")

# Verificar valores nulos
print(f"\nValores nulos:")
for col in columnas_interes:
    if col in f_tasa_reprobacion.columns:
        nulos = f_tasa_reprobacion[col].isnull().sum()
        porcentaje = (nulos / len(f_tasa_reprobacion)) * 100
        print(f"  • {col}: {nulos} ({porcentaje:.1f}%)")
```

```python
# 2. ESTADÍSTICAS DESCRIPTIVAS
print("\n" + "="*60)
print("ESTADÍSTICAS DESCRIPTIVAS")
print("="*60)

# Estadísticas para las tres variables de interés
estadisticas = f_tasa_reprobacion[['REPROBADOS', 'REGISTRADOS', 'TASA_RP']].describe()
print("\nEstadísticas descriptivas:")
print(estadisticas.round(2))

# Estadísticas adicionales
print(f"\n📈 ANÁLISIS DETALLADO POR VARIABLE:")

for col in ['REPROBADOS', 'REGISTRADOS', 'TASA_RP']:
    print(f"\n--- {col} ---")
    serie = f_tasa_reprobacion[col]
    print(f"• Media: {serie.mean():.2f}")
    print(f"• Mediana: {serie.median():.2f}")
    print(f"• Moda: {serie.mode().iloc[0] if not serie.mode().empty else 'No definida'}")
    print(f"• Desviación estándar: {serie.std():.2f}")
    print(f"• Coeficiente de variación: {(serie.std()/serie.mean())*100:.1f}%")
    print(f"• Rango: {serie.min()} - {serie.max()}")
    print(f"• Rango intercuartílico (Q3-Q1): {serie.quantile(0.75) - serie.quantile(0.25):.2f}")

# Información adicional
print(f"\n🔍 MÉTRICAS CALCULADAS:")
# Validar consistencia: TASA_RP debería ser REPROBADOS/REGISTRADOS
f_tasa_reprobacion['TASA_CALCULADA'] = f_tasa_reprobacion['REPROBADOS'] / f_tasa_reprobacion['REGISTRADOS']
diferencia_tasas = (f_tasa_reprobacion['TASA_RP'] - f_tasa_reprobacion['TASA_CALCULADA']).abs()

print(f"• Validación TASA_RP = REPROBADOS/REGISTRADOS:")
print(f"  - Diferencia máxima: {diferencia_tasas.max():.6f}")
print(f"  - Diferencia promedio: {diferencia_tasas.mean():.6f}")
if diferencia_tasas.max() < 0.001:
    print("  ✓ TASA_RP coincide con el cálculo esperado")
else:
    print("  ⚠️ Hay discrepancias en el cálculo de TASA_RP")

# Materias con más estudiantes registrados
print(f"\n📚 MATERIAS CON MÁS ESTUDIANTES:")
top_registrados = f_tasa_reprobacion.nlargest(5, 'REGISTRADOS')[['MATERIA', 'REGISTRADOS', 'REPROBADOS', 'TASA_RP']]
for _, row in top_registrados.iterrows():
    print(f"• {row['MATERIA'][:40]}: {row['REGISTRADOS']} registrados, {row['REPROBADOS']} reprobados ({row['TASA_RP']:.1%})")

# Materias con mayor tasa de reprobación
print(f"\n⚠️ MATERIAS CON MAYOR TASA DE REPROBACIÓN:")
top_tasa = f_tasa_reprobacion.nlargest(5, 'TASA_RP')[['MATERIA', 'REGISTRADOS', 'REPROBADOS', 'TASA_RP']]
for _, row in top_tasa.iterrows():
    print(f"• {row['MATERIA'][:40]}: {row['TASA_RP']:.1%} ({row['REPROBADOS']}/{row['REGISTRADOS']})")

# Materias con menor tasa de reprobación
print(f"\n✅ MATERIAS CON MENOR TASA DE REPROBACIÓN:")
top_aprobacion = f_tasa_reprobacion.nsmallest(5, 'TASA_RP')[['MATERIA', 'REGISTRADOS', 'REPROBADOS', 'TASA_RP']]
for _, row in top_aprobacion.iterrows():
    print(f"• {row['MATERIA'][:40]}: {row['TASA_RP']:.1%} ({row['REPROBADOS']}/{row['REGISTRADOS']})")
```

```python
# 3. VISUALIZACIÓN DE DISTRIBUCIONES
print("\n" + "="*60)
print("ANÁLISIS DE DISTRIBUCIONES")
print("="*60)

# Crear visualizaciones
plt.figure(figsize=(18, 12))

# 1. Histogramas de las tres variables
plt.subplot(3, 3, 1)
plt.hist(f_tasa_reprobacion['REGISTRADOS'], bins=30, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Distribución de REGISTRADOS')
plt.xlabel('Número de Registrados')
plt.ylabel('Frecuencia')
plt.grid(True, alpha=0.3)

plt.subplot(3, 3, 2)
plt.hist(f_tasa_reprobacion['REPROBADOS'], bins=30, alpha=0.7, color='salmon', edgecolor='black')
plt.title('Distribución de REPROBADOS')
plt.xlabel('Número de Reprobados')
plt.ylabel('Frecuencia')
plt.grid(True, alpha=0.3)

plt.subplot(3, 3, 3)
plt.hist(f_tasa_reprobacion['TASA_RP'], bins=30, alpha=0.7, color='lightgreen', edgecolor='black')
plt.title('Distribución de TASA_RP')
plt.xlabel('Tasa de Reprobación')
plt.ylabel('Frecuencia')
plt.grid(True, alpha=0.3)

# 2. Box plots
plt.subplot(3, 3, 4)
plt.boxplot(f_tasa_reprobacion['REGISTRADOS'])
plt.title('Box Plot - REGISTRADOS')
plt.ylabel('Número de Registrados')
plt.grid(True, alpha=0.3)

plt.subplot(3, 3, 5)
plt.boxplot(f_tasa_reprobacion['REPROBADOS'])
plt.title('Box Plot - REPROBADOS')
plt.ylabel('Número de Reprobados')
plt.grid(True, alpha=0.3)

plt.subplot(3, 3, 6)
plt.boxplot(f_tasa_reprobacion['TASA_RP'])
plt.title('Box Plot - TASA_RP')
plt.ylabel('Tasa de Reprobación')
plt.grid(True, alpha=0.3)

# 3. Scatter plots para relaciones
plt.subplot(3, 3, 7)
plt.scatter(f_tasa_reprobacion['REGISTRADOS'], f_tasa_reprobacion['REPROBADOS'], 
           alpha=0.6, color='purple', s=30)
plt.title('REGISTRADOS vs REPROBADOS')
plt.xlabel('Registrados')
plt.ylabel('Reprobados')
plt.grid(True, alpha=0.3)

plt.subplot(3, 3, 8)
plt.scatter(f_tasa_reprobacion['REGISTRADOS'], f_tasa_reprobacion['TASA_RP'], 
           alpha=0.6, color='orange', s=30)
plt.title('REGISTRADOS vs TASA_RP')
plt.xlabel('Registrados')
plt.ylabel('Tasa de Reprobación')
plt.grid(True, alpha=0.3)

plt.subplot(3, 3, 9)
plt.scatter(f_tasa_reprobacion['REPROBADOS'], f_tasa_reprobacion['TASA_RP'], 
           alpha=0.6, color='red', s=30)
plt.title('REPROBADOS vs TASA_RP')
plt.xlabel('Reprobados')
plt.ylabel('Tasa de Reprobación')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Análisis de correlaciones
print(f"\n🔗 CORRELACIONES ENTRE VARIABLES:")
correlaciones = f_tasa_reprobacion[['REGISTRADOS', 'REPROBADOS', 'TASA_RP']].corr()
print(correlaciones.round(3))

# Interpretación de correlaciones
print(f"\n📊 INTERPRETACIÓN DE CORRELACIONES:")
corr_reg_rep = correlaciones.loc['REGISTRADOS', 'REPROBADOS']
corr_reg_tasa = correlaciones.loc['REGISTRADOS', 'TASA_RP']
corr_rep_tasa = correlaciones.loc['REPROBADOS', 'TASA_RP']

print(f"• REGISTRADOS vs REPROBADOS: {corr_reg_rep:.3f}")
if corr_reg_rep > 0.7:
    print("  → Correlación fuerte positiva: a más registrados, más reprobados")
elif corr_reg_rep > 0.3:
    print("  → Correlación moderada positiva")
else:
    print("  → Correlación débil")

print(f"• REGISTRADOS vs TASA_RP: {corr_reg_tasa:.3f}")
if abs(corr_reg_tasa) < 0.3:
    print("  → Correlación débil: el tamaño de la materia no determina la dificultad")
elif corr_reg_tasa > 0:
    print("  → Las materias con más estudiantes tienden a tener mayor reprobación")
else:
    print("  → Las materias con más estudiantes tienden a tener menor reprobación")

print(f"• REPROBADOS vs TASA_RP: {corr_rep_tasa:.3f}")
if corr_rep_tasa > 0.3:
    print("  → A más reprobados absolutos, mayor tasa de reprobación")
else:
    print("  → El número absoluto de reprobados no determina la tasa")
```

```python
# 4. ANÁLISIS DE CATEGORIZACIÓN
print("\n" + "="*60)
print("ANÁLISIS DE CATEGORIZACIÓN POR TASA DE REPROBACIÓN")
print("="*60)

# Categorizar materias por tasa de reprobación
def categorizar_dificultad(tasa):
    if tasa <= 10:  # 10%
        return 'Muy Fácil'
    elif tasa <= 20:  # 20%
        return 'Fácil'
    elif tasa <= 30:  # 30%
        return 'Moderada'
    elif tasa <= 40:  # 40%
        return 'Difícil'
    else:
        return 'Muy Difícil'

f_tasa_reprobacion['CATEGORIA_DIFICULTAD'] = f_tasa_reprobacion['TASA_RP'].apply(categorizar_dificultad)

# Análisis por categorías
print("\n📊 DISTRIBUCIÓN POR CATEGORÍA DE DIFICULTAD:")
distribucion_categorias = f_tasa_reprobacion['CATEGORIA_DIFICULTAD'].value_counts()
print(distribucion_categorias)

# Porcentajes
print("\nPorcentajes:")
porcentajes_cat = f_tasa_reprobacion['CATEGORIA_DIFICULTAD'].value_counts(normalize=True) * 100
for categoria, porcentaje in porcentajes_cat.items():
    print(f"• {categoria}: {porcentaje:.1f}%")

# Estadísticas por categoría
print(f"\n📈 ESTADÍSTICAS POR CATEGORÍA:")
stats_por_categoria = f_tasa_reprobacion.groupby('CATEGORIA_DIFICULTAD')[['REGISTRADOS', 'REPROBADOS', 'TASA_RP']].agg(['mean', 'median', 'std'])
print(stats_por_categoria.round(2))

# Visualización por categorías
plt.figure(figsize=(15, 10))

# 1. Distribución de categorías
plt.subplot(2, 3, 1)
distribucion_categorias.plot(kind='bar', color=['green', 'lightgreen', 'yellow', 'orange', 'red'], alpha=0.7)
plt.title('Distribución por Categoría de Dificultad')
plt.xlabel('Categoría')
plt.ylabel('Número de Materias')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

# 2. Box plot de REGISTRADOS por categoría
plt.subplot(2, 3, 2)
categorias_orden = ['Muy Fácil', 'Fácil', 'Moderada', 'Difícil', 'Muy Difícil']
data_registrados = [f_tasa_reprobacion[f_tasa_reprobacion['CATEGORIA_DIFICULTAD']==cat]['REGISTRADOS'].values 
                   for cat in categorias_orden if cat in f_tasa_reprobacion['CATEGORIA_DIFICULTAD'].values]
labels_registrados = [cat for cat in categorias_orden if cat in f_tasa_reprobacion['CATEGORIA_DIFICULTAD'].values]
plt.boxplot(data_registrados, labels=labels_registrados)
plt.title('REGISTRADOS por Categoría')
plt.ylabel('Número de Registrados')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

# 3. Box plot de REPROBADOS por categoría
plt.subplot(2, 3, 3)
data_reprobados = [f_tasa_reprobacion[f_tasa_reprobacion['CATEGORIA_DIFICULTAD']==cat]['REPROBADOS'].values 
                  for cat in categorias_orden if cat in f_tasa_reprobacion['CATEGORIA_DIFICULTAD'].values]
plt.boxplot(data_reprobados, labels=labels_registrados)
plt.title('REPROBADOS por Categoría')
plt.ylabel('Número de Reprobados')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

# 4. Promedio de registrados por categoría
plt.subplot(2, 3, 4)
avg_registrados = f_tasa_reprobacion.groupby('CATEGORIA_DIFICULTAD')['REGISTRADOS'].mean()
avg_registrados.plot(kind='bar', color='skyblue', alpha=0.7)
plt.title('Promedio de REGISTRADOS por Categoría')
plt.ylabel('Promedio de Registrados')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

# 5. Promedio de reprobados por categoría
plt.subplot(2, 3, 5)
avg_reprobados = f_tasa_reprobacion.groupby('CATEGORIA_DIFICULTAD')['REPROBADOS'].mean()
avg_reprobados.plot(kind='bar', color='salmon', alpha=0.7)
plt.title('Promedio de REPROBADOS por Categoría')
plt.ylabel('Promedio de Reprobados')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

# 6. Dispersión REGISTRADOS vs TASA_RP coloreada por categoría
plt.subplot(2, 3, 6)
colores_categoria = {'Muy Fácil': 'green', 'Fácil': 'lightgreen', 'Moderada': 'yellow', 
                    'Difícil': 'orange', 'Muy Difícil': 'red'}
for categoria in f_tasa_reprobacion['CATEGORIA_DIFICULTAD'].unique():
    subset = f_tasa_reprobacion[f_tasa_reprobacion['CATEGORIA_DIFICULTAD'] == categoria]
    plt.scatter(subset['REGISTRADOS'], subset['TASA_RP'], 
               c=colores_categoria.get(categoria, 'gray'), label=categoria, alpha=0.6)
plt.xlabel('Registrados')
plt.ylabel('Tasa de Reprobación')
plt.title('REGISTRADOS vs TASA_RP por Categoría')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Identificar materias extremas
print(f"\n🔍 MATERIAS EN CATEGORÍAS EXTREMAS:")

print(f"\n✅ MATERIAS MUY FÁCILES (Tasa ≤ 10%):")
muy_faciles = f_tasa_reprobacion[f_tasa_reprobacion['CATEGORIA_DIFICULTAD'] == 'Muy Fácil'].nsmallest(5, 'TASA_RP')
for _, row in muy_faciles.iterrows():
    print(f"• {row['MATERIA'][:50]}: {row['TASA_RP']:.1%} ({row['REPROBADOS']}/{row['REGISTRADOS']})")

print(f"\n⚠️ MATERIAS MUY DIFÍCILES (Tasa > 40%):")
muy_dificiles = f_tasa_reprobacion[f_tasa_reprobacion['CATEGORIA_DIFICULTAD'] == 'Muy Difícil'].nlargest(5, 'TASA_RP')
for _, row in muy_dificiles.iterrows():
    print(f"• {row['MATERIA'][:50]}: {row['TASA_RP']:.1%} ({row['REPROBADOS']}/{row['REGISTRADOS']})")
```

```python
# 5. INVESTIGACIÓN DE ANOMALÍAS EN LOS DATOS
print("\n" + "="*60)
print("INVESTIGACIÓN DE ANOMALÍAS - TASAS > 100%")
print("="*60)

# Identificar registros con tasa > 100 (para valores en escala 0-100)
tasas_anomalas = f_tasa_reprobacion[f_tasa_reprobacion['TASA_RP'] > 100.0]
print(f"🚨 REGISTROS CON TASA DE REPROBACIÓN > 100%:")
print(f"• Número de registros anómalos: {len(tasas_anomalas)}")
print(f"• Porcentaje del total: {len(tasas_anomalas)/len(f_tasa_reprobacion)*100:.1f}%")

if len(tasas_anomalas) > 0:
    print(f"\n🔍 ANÁLISIS DE CASOS ANÓMALOS:")
    print("Top 10 casos con mayor tasa:")
    anomalos_top = tasas_anomalas.nlargest(10, 'TASA_RP')[['MATERIA', 'REGISTRADOS', 'REPROBADOS', 'TASA_RP']]
    for _, row in anomalos_top.iterrows():
        print(f"• {row['MATERIA'][:40]}: {row['REPROBADOS']}/{row['REGISTRADOS']} = {row['TASA_RP']:.1f}%")
    
    print(f"\n📊 ESTADÍSTICAS DE CASOS ANÓMALOS:")
    print(f"• TASA_RP mínima en casos anómalos: {tasas_anomalas['TASA_RP'].min():.1f}%")
    print(f"• TASA_RP máxima en casos anómalos: {tasas_anomalas['TASA_RP'].max():.1f}%")
    print(f"• TASA_RP promedio en casos anómalos: {tasas_anomalas['TASA_RP'].mean():.1f}%")
    
    # Verificar si el problema es en el cálculo
    print(f"\n🔧 VERIFICACIÓN DE CÁLCULOS:")
    print("Comparación TASA_RP vs (REPROBADOS/REGISTRADOS)*100 para casos anómalos:")
    tasas_anomalas_sample = tasas_anomalas.head(5).copy()
    tasas_anomalas_sample['TASA_CALCULADA_MANUAL'] = (tasas_anomalas_sample['REPROBADOS'] / tasas_anomalas_sample['REGISTRADOS']) * 100
    tasas_anomalas_sample['DIFERENCIA'] = tasas_anomalas_sample['TASA_RP'] - tasas_anomalas_sample['TASA_CALCULADA_MANUAL']
    
    print(tasas_anomalas_sample[['MATERIA', 'REPROBADOS', 'REGISTRADOS', 'TASA_RP', 'TASA_CALCULADA_MANUAL', 'DIFERENCIA']])

# Análisis de distribución normal (sin anomalías)
tasas_normales = f_tasa_reprobacion[f_tasa_reprobacion['TASA_RP'] <= 100.0]
print(f"\n📈 ANÁLISIS DE DATOS NORMALES (TASA ≤ 100%):")
print(f"• Número de registros normales: {len(tasas_normales)}")
print(f"• Porcentaje del total: {len(tasas_normales)/len(f_tasa_reprobacion)*100:.1f}%")

if len(tasas_normales) > 0:
    print(f"\nEstadísticas de tasas normales:")
    print(f"• Promedio: {tasas_normales['TASA_RP'].mean():.1f}%")
    print(f"• Mediana: {tasas_normales['TASA_RP'].median():.1f}%")
    print(f"• Desviación estándar: {tasas_normales['TASA_RP'].std():.1f}")
    print(f"• Rango: {tasas_normales['TASA_RP'].min():.1f}% - {tasas_normales['TASA_RP'].max():.1f}%")
    
    # Recategorizar solo datos normales
    tasas_normales_cat = tasas_normales.copy()
    tasas_normales_cat['CATEGORIA_DIFICULTAD_CORREGIDA'] = tasas_normales_cat['TASA_RP'].apply(categorizar_dificultad)
    
    print(f"\n📊 DISTRIBUCIÓN CORREGIDA POR CATEGORÍA DE DIFICULTAD (solo datos normales):")
    distribucion_corregida = tasas_normales_cat['CATEGORIA_DIFICULTAD_CORREGIDA'].value_counts()
    print(distribucion_corregida)
    
    # Porcentajes corregidos
    print("\nPorcentajes corregidos:")
    porcentajes_corregidos = tasas_normales_cat['CATEGORIA_DIFICULTAD_CORREGIDA'].value_counts(normalize=True) * 100
    for categoria, porcentaje in porcentajes_corregidos.items():
        print(f"• {categoria}: {porcentaje:.1f}%")

# Recomendación sobre el tratamiento de datos
print(f"\n💡 RECOMENDACIONES SOBRE CALIDAD DE DATOS:")
print("• Los casos con tasa > 100% sugieren posibles errores en los datos fuente")
print("• Posibles causas:")
print("  - Error en el conteo de reprobados vs registrados")
print("  - Datos de diferentes períodos mezclados incorrectamente")
print("  - Problemas en la fuente de datos original")
print("• Para análisis futuros, considerar:")
print("  - Filtrar casos con TASA_RP <= 100.0")
print("  - Recalcular TASA_RP como (REPROBADOS/REGISTRADOS)*100")
print("  - Verificar la fuente y metodología de cálculo original")
```

```python
# Conclusiones y recomendaciones del análisis
print("\n" + "="*80)
print("CONCLUSIONES Y RECOMENDACIONES")
print("="*80)

# Análisis de diferencias entre metodologías
metodologias = {
    'Aprobadas': 'Solo estudiantes que aprobaron la materia',
    'Aprobadas_Reprobadas': 'Incluye estudiantes aprobados y reprobados',
    'Aprobadas_SinImportarVeces': 'Solo aprobados, sin importar intentos',
    'Aprobadas_Reprobadas_SinImportarVeces': 'Todos los estudiantes, sin importar intentos'
}

print("📊 DIFERENCIAS METODOLÓGICAS:")
for nombre, descripcion in metodologias.items():
    df_temp = datasets[nombre]
    print(f"\n• {nombre}:")
    print(f"  - Descripción: {descripcion}")
    print(f"  - Materias analizadas: {len(df_temp)}")
    print(f"  - DIFICULTAD promedio: {df_temp['DIFICULTAD'].mean():.3f}")
    print(f"  - Rango DIFICULTAD: {df_temp['DIFICULTAD'].min():.3f} - {df_temp['DIFICULTAD'].max():.3f}")

print(f"\n🔍 HALLAZGOS PRINCIPALES:")

# Identificar el dataset más conservador y más liberal
gpa_promedios = {nombre: df['DIFICULTAD'].mean() for nombre, df in datasets.items()}
dataset_conservador = min(gpa_promedios, key=gpa_promedios.get)
dataset_liberal = max(gpa_promedios, key=gpa_promedios.get)

print(f"• Dataset más conservador (menor DIFICULTAD promedio): {dataset_conservador}")
print(f"  - DIFICULTAD promedio: {gpa_promedios[dataset_conservador]:.3f}")
print(f"• Dataset más liberal (mayor DIFICULTAD promedio): {dataset_liberal}")
print(f"  - DIFICULTAD promedio: {gpa_promedios[dataset_liberal]:.3f}")

# Diferencia entre metodologías
diferencia_metodologias = gpa_promedios[dataset_liberal] - gpa_promedios[dataset_conservador]
print(f"• Diferencia entre metodologías: {diferencia_metodologias:.3f} puntos de DIFICULTAD")

print(f"\n💡 INTERPRETACIONES:")
print("• Incluir estudiantes reprobados REDUCE el DIFICULTAD promedio de las materias")
print("• Considerar múltiples intentos puede AUMENTAR o REDUCIR el DIFICULTAD dependiendo del criterio")
print("• Las diferencias metodológicas son significativas para la evaluación de dificultad")

print(f"\n📋 RECOMENDACIONES:")
print("• Para identificar materias realmente difíciles: usar dataset 'Aprobadas_Reprobadas'")
print("• Para ranking de materias por rendimiento: usar dataset 'Aprobadas'") 
print("• Para análisis longitudinal: considerar datasets 'SinImportarVeces'")
print("• Combinar múltiples métricas para una evaluación integral")
print("• Considerar el contexto específico del análisis al elegir la metodología")

# Crear un DataFrame resumen final
resumen_final = pd.DataFrame({
    'Metodologia': list(datasets.keys()),
    'Num_Materias': [len(df) for df in datasets.values()],
    'DIFICULTAD_Promedio': [df['DIFICULTAD'].mean() for df in datasets.values()],
    'DIFICULTAD_StdDev': [df['DIFICULTAD'].std() for df in datasets.values()],
    'Materia_Mas_Dificil': [df.loc[df['DIFICULTAD'].idxmin(), 'MATERIA'][:30] for df in datasets.values()],
    'DIFICULTAD_Mas_Bajo': [df['DIFICULTAD'].min() for df in datasets.values()]
})

print(f"\n📈 RESUMEN COMPARATIVO FINAL:")
print(resumen_final.round(3))
```
