# ANÁLISIS EXPLORATORIO DE DATOS - RIESGO ACADÉMICO

En este análisis exploraremos los datos de la materia **MATG1076** para entender los factores que influyen en la reprobación académica y las características socioeconómicas de los estudiantes.

```python
# 1. INFORMACIÓN GENERAL DEL DATASET
print("="*60)
print("INFORMACIÓN GENERAL DEL DATASET")
print("="*60)
print(f"Dimensiones del dataset: {df.shape}")
print(f"Número de estudiantes: {df.shape[0]}")
print(f"Número de variables: {df.shape[1]}")
print(f"Código de materia analizada: {cod_materia}")
print("\nTipos de datos:")
print(df.dtypes.value_counts())
print("\nValores nulos por columna:")
valores_nulos = df.isnull().sum()
print(f"Total de columnas con valores nulos: {(valores_nulos > 0).sum()}")
if (valores_nulos > 0).sum() > 0:
    print("\nColumnas con valores nulos:")
    print(valores_nulos[valores_nulos > 0])
```

```python
# 2. ANÁLISIS DE LA VARIABLE OBJETIVO (REPROBACIÓN)
print("="*60)
print("ANÁLISIS DE LA VARIABLE OBJETIVO")
print("="*60)

# Identificar columnas relacionadas con reprobación
columnas_reprobacion = [col for col in df.columns if 'reprob' in col.lower() or 'aprobo' in col.lower() or 'estado' in col.lower()]
print(f"Posibles columnas de variable objetivo: {columnas_reprobacion}")

# Si encontramos columnas relacionadas, analizarlas
if columnas_reprobacion:
    for col in columnas_reprobacion[:3]:  # Analizar las primeras 3
        print(f"\nAnálisis de {col}:")
        print(df[col].value_counts())
        print(f"Porcentaje de distribución:")
        print(df[col].value_counts(normalize=True) * 100)

# Visualización de la distribución
plt.figure(figsize=(12, 4))
for i, col in enumerate(columnas_reprobacion[:2], 1):
    plt.subplot(1, 2, i)
    df[col].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
    plt.title(f'Distribución de {col}')
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 3. ANÁLISIS DE VARIABLES NUMÉRICAS
print("="*60)
print("ANÁLISIS DE VARIABLES NUMÉRICAS")
print("="*60)

# Identificar variables numéricas
variables_numericas = df.select_dtypes(include=[np.number]).columns.tolist()
print(f"Variables numéricas encontradas: {len(variables_numericas)}")
print(variables_numericas[:10])  # Mostrar las primeras 10

# Estadísticas descriptivas
print("\nEstadísticas descriptivas de variables numéricas principales:")
print(df[variables_numericas[:8]].describe().round(2))

# Visualización de distribuciones
plt.figure(figsize=(15, 12))
for i, var in enumerate(variables_numericas[:12], 1):
    plt.subplot(3, 4, i)
    df[var].hist(bins=20, alpha=0.7, color='skyblue', edgecolor='black')
    plt.title(f'Distribución de {var}')
    plt.xlabel(var)
    plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()
```

```python
# 4. ANÁLISIS DE VARIABLES CATEGÓRICAS
print("="*60)
print("ANÁLISIS DE VARIABLES CATEGÓRICAS")
print("="*60)

# Identificar variables categóricas
variables_categoricas = df.select_dtypes(include=['object', 'category']).columns.tolist()
print(f"Variables categóricas encontradas: {len(variables_categoricas)}")
print(variables_categoricas[:10])  # Mostrar las primeras 10

# Análisis de cardinalidad (número de valores únicos)
print("\nCardinalidad de variables categóricas:")
for var in variables_categoricas[:10]:
    unique_vals = df[var].nunique()
    print(f"{var}: {unique_vals} valores únicos")
    if unique_vals <= 10:  # Mostrar valores únicos si son pocos
        print(f"  Valores: {df[var].unique()}")
    print()

# Visualización de variables categóricas con pocas categorías
variables_cat_simples = [var for var in variables_categoricas if df[var].nunique() <= 6][:8]
if len(variables_cat_simples) > 0:
    plt.figure(figsize=(15, 8))
    for i, var in enumerate(variables_cat_simples, 1):
        plt.subplot(2, 4, i)
        df[var].value_counts().plot(kind='bar', color='lightgreen')
        plt.title(f'Distribución de {var}')
        plt.xticks(rotation=45)
        plt.ylabel('Frecuencia')
    plt.tight_layout()
    plt.show()
```

```python
# 2. Correlación entre variables numéricas
# Visualizamos la matriz de correlación
corr_matrix = df.select_dtypes(include=['int64', 'float64']).corr()

# Crear un mapa de calor (heatmap)
plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))  # Máscara para mostrar solo la mitad inferior
sns_heatmap = plt.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(sns_heatmap)
plt.title('Matriz de Correlación de Variables Numéricas', fontsize=16)
plt.xticks(np.arange(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(np.arange(len(corr_matrix.columns)), corr_matrix.columns)
plt.tight_layout()
plt.show()

# Mostrar las correlaciones más fuertes con la variable objetivo 'pass'
if 'pass' in corr_matrix.columns:
    correlaciones_pass = corr_matrix['pass'].sort_values(ascending=False)
    print("\nCorrelaciones con la variable 'pass' (aprobación):")
    print(correlaciones_pass)
```

```python
# 5. ANÁLISIS DE CORRELACIONES
print("="*60)
print("ANÁLISIS DE CORRELACIONES")
print("="*60)

# Matriz de correlación para variables numéricas
variables_numericas = df.select_dtypes(include=[np.number]).columns.tolist()
if len(variables_numericas) > 1:
    corr_matrix = df[variables_numericas].corr()
    
    # Ajustar el tamaño de la figura según el número de variables
    n_vars = len(variables_numericas)
    fig_size = max(12, n_vars * 0.8)  # Tamaño mínimo de 12, escalable
    
    # # Visualización de la matriz de correlación completa
    # plt.figure(figsize=(fig_size, fig_size))
    # mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
    
    # # Ajustar parámetros según el número de variables
    # if n_vars > 15:
    #     # Para muchas variables, no mostrar anotaciones
    #     sns.heatmap(corr_matrix, mask=mask, cmap='coolwarm', center=0,
    #                 square=True, linewidths=0.2, cbar_kws={"shrink": .8})
    #     plt.title('Matriz de Correlación - Variables Numéricas', fontsize=16)
    # else:
    #     # Para pocas variables, mostrar anotaciones
    #     sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='coolwarm', center=0,
    #                 square=True, linewidths=0.5, cbar_kws={"shrink": .8}, fmt='.2f',
    #                 annot_kws={'size': 8})
    #     plt.title('Matriz de Correlación - Variables Numéricas', fontsize=16)
    
    # # Rotar etiquetas para mejor legibilidad
    # plt.xticks(rotation=45, ha='right', fontsize=10)
    # plt.yticks(rotation=0, fontsize=10)
    # plt.tight_layout()
    # plt.show()
    
    # Mostrar matriz más pequeña con variables más importantes si hay muchas
    if n_vars > 15:
        print(f"\nMatriz muy grande ({n_vars} variables). Mostrando matriz reducida con variables más importantes:")
        # Seleccionar variables con mayor varianza para una matriz reducida
        vars_importantes = variables_numericas[:12]  # Primeras 12 variables
        corr_reducida = df[vars_importantes].corr()
        
        plt.figure(figsize=(12, 10))
        mask_red = np.triu(np.ones_like(corr_reducida, dtype=bool))
        sns.heatmap(corr_reducida, mask=mask_red, annot=True, cmap='coolwarm', center=0,
                    square=True, linewidths=0.5, cbar_kws={"shrink": .8}, fmt='.2f')
        plt.title('Matriz de Correlación - Variables Principales', fontsize=14)
        plt.xticks(rotation=45, ha='right')
        plt.yticks(rotation=0)
        plt.tight_layout()
        plt.show()
    
    # Encontrar correlaciones más fuertes
    print("Correlaciones más fuertes (|r| > 0.7):")
    high_corr = np.where(np.abs(corr_matrix) > 0.7)
    correlaciones_fuertes = []
    for i, j in zip(high_corr[0], high_corr[1]):
        if i < j:  # Evitar duplicados
            var1, var2 = corr_matrix.index[i], corr_matrix.columns[j]
            corr_val = corr_matrix.iloc[i, j]
            correlaciones_fuertes.append((var1, var2, corr_val))
            print(f"{var1} - {var2}: {corr_val:.3f}")
    
    if not correlaciones_fuertes:
        print("No se encontraron correlaciones fuertes (|r| > 0.7)")
        print("Correlaciones moderadas (|r| > 0.5):")
        high_corr_mod = np.where(np.abs(corr_matrix) > 0.5)
        for i, j in zip(high_corr_mod[0], high_corr_mod[1]):
            if i < j:
                var1, var2 = corr_matrix.index[i], corr_matrix.columns[j]
                corr_val = corr_matrix.iloc[i, j]
                print(f"{var1} - {var2}: {corr_val:.3f}")
                
else:
    print("No hay suficientes variables numéricas para análisis de correlación")
```

```python
# 6. ANÁLISIS SOCIOECONÓMICO
print("="*60)
print("ANÁLISIS DE VARIABLES SOCIOECONÓMICAS")
print("="*60)

# Identificar variables socioeconómicas
variables_socioeconomicas = [col for col in df.columns if any(keyword in col.lower() 
                            for keyword in ['socio', 'econ', 'ingreso', 'estrato', 'nivel', 'beca', 'subsidio'])]
print(f"Variables socioeconómicas identificadas: {variables_socioeconomicas}")

# Análisis específico de variables socioeconómicas
for var in variables_socioeconomicas[:5]:  # Analizar las primeras 5
    print(f"\nAnálisis de {var}:")
    if df[var].dtype in ['object', 'category']:
        print(df[var].value_counts())
        print(f"Porcentajes:")
        print((df[var].value_counts(normalize=True) * 100).round(2))
    else:
        print(df[var].describe())

# Visualización de variables socioeconómicas
if len(variables_socioeconomicas) > 0:
    plt.figure(figsize=(15, 8))
    for i, var in enumerate(variables_socioeconomicas[:6], 1):
        plt.subplot(2, 3, i)
        if df[var].dtype in ['object', 'category'] and df[var].nunique() <= 10:
            df[var].value_counts().plot(kind='bar', color='orange', alpha=0.7)
            plt.xticks(rotation=45)
        else:
            df[var].hist(bins=15, alpha=0.7, color='orange')
        plt.title(f'Distribución de {var}')
        plt.ylabel('Frecuencia')
    plt.tight_layout()
    plt.show()
```

```python
# 7. RELACIÓN ENTRE VARIABLES Y REPROBACIÓN
print("="*60)
print("RELACIÓN VARIABLES VS REPROBACIÓN")
print("="*60)

# Encontrar la variable de reprobación/resultado académico
columnas_resultado = [col for col in df.columns if any(keyword in col.lower() 
                     for keyword in ['reprob', 'aprobo', 'resultado', 'estado', 'calificacion'])]

if columnas_resultado:
    var_objetivo = columnas_resultado[0]  # Tomar la primera
    print(f"Variable objetivo seleccionada: {var_objetivo}")
    print(f"Valores únicos: {df[var_objetivo].unique()}")
    
    # CONVERSIÓN A NUMÉRICO PARA CORRELACIONES
    # Crear una copia de la variable objetivo convertida a numérico
    if df[var_objetivo].dtype == 'object':
        # Mapear valores categóricos a numéricos
        if set(df[var_objetivo].unique()).issubset({'AP', 'RP'}):
            df_temp = df.copy()
            df_temp[f'{var_objetivo}_num'] = df_temp[var_objetivo].map({'AP': 1, 'RP': 0})
            var_objetivo_num = f'{var_objetivo}_num'
            print(f"Conversión: AP=1 (Aprobado), RP=0 (Reprobado)")
        else:
            # Para otros casos, usar factorize
            df_temp = df.copy()
            df_temp[f'{var_objetivo}_num'] = pd.factorize(df_temp[var_objetivo])[0]
            var_objetivo_num = f'{var_objetivo}_num'
            print(df_temp[var_objetivo].unique())
            print(df_temp[var_objetivo_num].unique())
            print(f"Conversión numérica automática aplicada")
    else:
        df_temp = df.copy()
        var_objetivo_num = var_objetivo
    
    # Análisis bivariado con variables numéricas
    variables_numericas = df_temp.select_dtypes(include=[np.number]).columns.tolist()
    if var_objetivo_num in variables_numericas:
        variables_numericas.remove(var_objetivo_num)
    
    # Correlaciones con la variable objetivo (ahora numérica)
    if len(variables_numericas) > 0:
        print(f"\nCorrelaciones con {var_objetivo} (convertido a numérico):")
        correlaciones = df_temp[variables_numericas + [var_objetivo_num]].corr()[var_objetivo_num].sort_values(ascending=False)
        correlaciones_filtradas = correlaciones.drop(var_objetivo_num)
        print(correlaciones_filtradas)
        
        # Visualizar las correlaciones más importantes
        if len(correlaciones_filtradas) > 0:
            top_corr = correlaciones_filtradas.abs().sort_values(ascending=False)[:8]
            plt.figure(figsize=(12, 6))
            colors = ['red' if x < 0 else 'green' for x in correlaciones_filtradas[top_corr.index]]
            top_corr.plot(kind='bar', color=colors, alpha=0.7)
            plt.title(f'Correlaciones más fuertes con {var_objetivo}')
            plt.ylabel('Correlación (valor absoluto)')
            plt.xlabel('Variables')
            plt.xticks(rotation=45, ha='right')
            plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
            plt.tight_layout()
            plt.show()
    
    # Análisis categórico vs variable objetivo
    variables_categoricas = df.select_dtypes(include=['object', 'category']).columns.tolist()
    if var_objetivo in variables_categoricas:
        variables_categoricas.remove(var_objetivo)
        
    if len(variables_categoricas) > 0:
        print(f"\nAnálisis categórico vs {var_objetivo}:")
        n_vars = min(6, len(variables_categoricas))
        if n_vars > 0:
            plt.figure(figsize=(15, 10))
            for i, var_cat in enumerate(variables_categoricas[:n_vars], 1):
                plt.subplot(2, 3, i)
                tabla_cruzada = pd.crosstab(df[var_cat], df[var_objetivo], normalize='index') * 100
                tabla_cruzada.plot(kind='bar', ax=plt.gca(), color=['lightgreen', 'lightcoral'])
                plt.title(f'{var_cat} vs {var_objetivo}')
                plt.ylabel('Porcentaje')
                plt.xticks(rotation=90)
                plt.legend(title=var_objetivo)
            plt.tight_layout()
            plt.show()
            
            # Mostrar tabla de contingencia para la primera variable categórica
            print(f"\nTabla de contingencia - {variables_categoricas[0]} vs {var_objetivo}:")
            tabla = pd.crosstab(df[variables_categoricas[0]], df[var_objetivo], margins=True)
            print(tabla)
    
    # Estadísticas descriptivas por grupo
    print(f"\nEstadísticas descriptivas por {var_objetivo}:")
    for valor in df[var_objetivo].unique():
        print(f"\n--- {valor} ---")
        subset = df[df[var_objetivo] == valor]
        print(f"Cantidad: {len(subset)}")
        if len(variables_numericas) > 0:
            print("Promedios de variables numéricas:")
            promedios = subset[variables_numericas].mean()
            print(promedios.round(2))
            
else:
    print("No se encontró variable objetivo clara para el análisis")
```

```python
# 8. DETECCIÓN DE VALORES ATÍPICOS (OUTLIERS)
print("="*60)
print("DETECCIÓN DE VALORES ATÍPICOS")
print("="*60)

variables_numericas = df.select_dtypes(include=[np.number]).columns.tolist()

if len(variables_numericas) > 0:
    # Detectar outliers usando el método IQR
    outliers_summary = {}
    for var in variables_numericas[:8]:  # Analizar las primeras 8 variables
        Q1 = df[var].quantile(0.25)
        Q3 = df[var].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        outliers = df[(df[var] < lower_bound) | (df[var] > upper_bound)]
        outliers_summary[var] = len(outliers)
        
        print(f"{var}: {len(outliers)} outliers detectados ({len(outliers)/len(df)*100:.1f}%)")
    
    # Visualización de boxplots para detectar outliers
    plt.figure(figsize=(15, 10))
    for i, var in enumerate(variables_numericas[:8], 1):
        plt.subplot(2, 4, i)
        plt.boxplot(df[var].dropna())
        plt.title(f'Boxplot de {var}')
        plt.ylabel(var)
    plt.tight_layout()
    plt.show()
    
    # Resumen de outliers
    outliers_df = pd.DataFrame(list(outliers_summary.items()), 
                              columns=['Variable', 'Número_Outliers'])
    outliers_df['Porcentaje'] = (outliers_df['Número_Outliers'] / len(df)) * 100
    print("\nResumen de outliers por variable:")
    print(outliers_df.round(2))
```
